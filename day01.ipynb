find()
문자열 내부에 포함된 특정 문자열을 찾고자 할 때 사용  
찾고자하는 문자열이 있으면 그 문자열이 처음으로 나온 위치(인덱스)를 반환  
만일 찾고자 하는 문자열이 없는 경우 -1을 반환

인덱스를 이용해 검색 범위 지정 가능  
지정하지 않으면 처음부터 찾고,  
시작할 인덱스를 지정하는 경우에도 지정된 인덱스부터 찾음

rfind()메소드를 통해 오른쪽에서부터 값을 탐색 가능

```python
'apple watch is very expensive'.find('apple')
print('apple watch is very expensive'.find('a')) #인덱스 번호를 알려줌
print('apple watch is very expensive'.find('watch')) #시작 index 번호로 보여줌
print('apple watch is very expensive'.find('samsung')) #없으면 -1로 보여줌
print('apple watch is very expensive lets watch'.find('watch')) #시작 index 번호로 보여줌

#만약에 오른쪽에 왼쪽으로 찾고싶다면? rfind를 사용
print('apple watch is very expensive lets watch'.rfind('watch')) #시작 index 번호로 보여줌
```

대소문자 변환

|메소드|의미|
|------|----|
|upper()|모두 대문자로 변환한 결과를 반환|
|lower()|모두 소문자로 변환한 결과를 반환|
|capitalize()|첫 글자는 대문자로 변환하고 나머지 문자는 소문자로 변환한 결과 반환|

```python
print('apple'.upper()) #모두 대문자로변환
print('aPPLe'.upper()) #모두 대문자로변환

print('apple watch is very expensive'.capitalize())
print('apple watcH is very expensive'.capitalize())
```

join()
인수로 전달한 반복가능객체(문자열, 리스트 등)의 각 요소 사이에  
문자열을 포함시켜 새로운 문자열을 만들고 그 결과를 반환

```python
a= ['apple','watch','is','very','expensive']
full_sentense = ""
for i in a:
    full_sentense += (i + " ")
print(full_sentense)
```

```python
print(' '.join(['apple','watch','is','very','expensive']))
print('+'.join(['apple','watch','is','very','expensive']))
```

split()

하나의 문자열을 여러 개의 문자열로 분리 후 리스트로 반환하는 메소드  
기본적으로는 공백문자로 분리하지만 특정 문자열을 기준으로 분리 가능

```python
print('apple watch is very expensive'.split(" ")) #방법 1
print('apple watch is very expensive'.split()) #방법 2
```

불필요한 문자열 제거 메소드

|메소드|의미|
|--------|--------|
|lstrip  |왼쪽 끝에 있는 공백을 제거 후 반환|

if-else

if-else문은 조건식을 만족하는 경우와 만족하지 않는 경우를 
구분할 필요가 있을 시 사용.

사용법
* if 조건식:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 False일 때 실행문

조건식의 결과가 True인 경우와 False인 경우로 구분하여 코드 작성 가능.

if-elif

if-elif문은 특정 조건에 따라 3가지 이상으로 구분해야 할 때 사용.
여기서 elif는 else if

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문

if-elif-else

if-elif문은 특정 조건에 따라 구분 후 위의 조건을 만족하는 않는  
모든 경우에 대해 실행을 할 때 사용.

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2, 3이 모두 False일 때 실행될 문장.
