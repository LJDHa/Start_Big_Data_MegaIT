```python
def get_page_num_from_naver_news (oid,date,page):
    url = f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid={oid}&date={date}&page={page}'
    response_web_news = requests.get(url, headers = {'User-Agent': 'Mozilla 5.0'} )
    bs = BeautifulSoup(response_web_news.text,'lxml')
    last_page_num = bs.select('div.paging > strong')
    last_page_num = int(last_page_num[0].text)
    return last_page_num
```

```python
def get_bs_data_from_naver_news2(oid, date, last_page_num):
    bs_list_f =[]
    for page in range (1, last_page_num+1):
        url = f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid={oid}&date={date}&page={page}'
        response = requests.get(url, headers = {'User-Agent':'Mozilla 5.0'})
        bs= BeautifulSoup(response.text, 'lxml')
        bs_list = bs.select('table div#main_content dl a')
        bs_list_f += bs_list

    return bs_list_f
```

```python
def get_bs_data_from_naver_url2(oid, date, last_page_num):
    bs_url_f =[]
    for page in range (1, last_page_num+1):
        url = f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid={oid}&date={date}&page={page}'
        response = requests.get(url, headers = {'User-Agent':'Mozilla 5.0'})
        bs= BeautifulSoup(response.text, 'lxml')
        bs_url = bs.select('div dt a')
        bs_url_f += bs_url

    return bs_url_f
```
